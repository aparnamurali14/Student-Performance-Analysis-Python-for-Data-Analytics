```python
# Task 3: Overall Student Performance Analysis : Calculate:
#Average math score   
avg_math_score=data["math score"].mean()
print("The average math score=",avg_math_score)
#Average reading score
avg_reading_score=data["reading score"].mean()
print("The average reading score=",avg_reading_score)
#Average writing score
avg_writing_score=data["writing score"].mean()
print("The average writing score=",avg_writing_score)
```

```python
# 2.Identify:Highest score in each subject
highest_math_score=data["math score"].max()
highest_reading_score=data["reading score"].max()
highest_writing_score=data["writing score"].max()
print("Highest score in maths=",highest_math_score)
print("Highest score in reading=",highest_reading_score)
print("Highest score in writing=",highest_writing_score) 
```

```python
# 2.Identify:Lowest score in each subject
lowest_math_score=data["math score"].min()
lowest_reading_score=data["reading score"].min()
lowest_writing_score=data["writing score"].min()
print("Lowest score in maths=",lowest_math_score)
print("Lowest score in reading=",lowest_reading_score)
print("Lowest score in writing=",lowest_writing_score) 
```

```python
# 3. Find the total score for each student (sum of all three subjects).
total_score=(data["math score"]+data["reading score"]+data["writing score"])
```

```python
#  4. Add a new column total_score to the dataset.
data["Total score"]=total_score
data
```

```python
# Task 4: Gender-Based Performance Study
# 1.Calculate average scores (math, reading, writing) for each gender. 
data.groupby("gender")[["math score","reading score","writing score"]].mean()
```

```python
# 2.Identify: 
# Which gender performs better in math
avg_scores = data.groupby("gender")["math score"].mean()
print("Average math scores by gender:\n", avg_scores)
best_gender = avg_scores.idxmax()
print(f"{best_gender.capitalize()} performs better in math")
# Which gender performs better in reading and writing
avge_scores = data.groupby("gender")[["reading score", "writing score"]].mean()
print("\nAverage reading and writing scores by gender:\n", avge_scores)
best_gender = avge_scores.mean(axis=1).idxmax()
print(f"{best_gender.capitalize()} performs better in reading and writing")
```

```python
# 3.Display the performance comparison.
performance_comparison = data.groupby("gender")[["math score", "reading score", "writing score"]].mean()
print("\n     === Gender Performance Comparison ===")
print(performance_comparison.round(2))
```

```python
# �Task 5: Impact of Test Preparation Course The board invested in test preparation programs and wants to measure their impact. 
# 1. Separate students who: 
   # o Completed the test preparation course 
completed = data[data["test preparation course"] == "completed"]
   # o Did not complete the course 
not_completed = data[data["test preparation course"] == "none"]
# 2. Calculate average scores for both groups. 
completed_avg = completed[["math score", "reading score", "writing score"]].mean()
not_completed_avg = not_completed[["math score", "reading score", "writing score"]].mean()
print("Average Scores - Completed Course:\n", completed_avg)
print("\nAverage Scores - Did Not Complete Course:\n", not_completed_avg)
# 3. Compare performance across math, reading, and writing. 
comparison = data.groupby("test preparation course")[["math score", "reading score", "writing score"]].mean()
print("\n                  === Performance Comparison ===\n")
print(comparison)
# 4. Conclude whether test preparation 
if completed_avg.mean() > not_completed_avg.mean():
    print("\nConclusion: Students who completed the test preparation course performed better overall.")
else:
    print("\nConclusion: The test preparation course did not significantly improve performance.")
```

```python
# �Task 6: Parental Education & Student Performance The board believes parental education influences student outcomes. 
# 1. Group students by parental level of education. 
grouped = data.groupby("parental level of education")
# 2. Calculate average scores for each education level.
education_performance = grouped[["math score", "reading score", "writing score"]].mean()
print("            === Average Scores by Parental Education Level ===\n")
print(education_performance)
# 3. Identify:
# o Highest performing parental education group
overall_avg = education_performance.mean(axis=1)
highest_group = overall_avg.idxmax()
print("\nHighest Performing Parental Education Group:", highest_group)
# o Lowest performing group
lowest_group = overall_avg.idxmin()
print("\nLowest Performing Parental Education Group:", lowest_group)
```

```python
# �Task 7: Lunch Program & Academic Achievement The government provides different lunch programs to students. 
# 1. Analyze student performance based on lunch type. 
lunch_grouped = data.groupby("lunch")
# 2. Calculate average scores for each lunch category. 
lunch_performance = lunch_grouped[["math score", "reading score", "writing score"]].mean()
print("    === Average Scores by Lunch Type ===\n")
print(lunch_performance)
# 3. Identify which lunch program group performs better overall
overall_lunch_avg = lunch_performance.mean(axis=1)
best_lunch_group = overall_lunch_avg.idxmax()
print("\nHighest Performing Lunch Program Group:", best_lunch_group)
```

```python
# �Task 8: Performance Categorization (Data Manipulation).To simplify reporting, the board wants students categorized. 
# 1. Create a new column performance_level: 
# o Total Score ≥ 250 → Excellent
# o Total Score 200–249 → Good
# o Total Score < 200 → Needs Improvement
def categorize_performance(total_score):
    if total_score >= 250:
        return "Excellent"
    elif 200 <= total_score < 250:
        return "Good"
    else:
        return "Needs Improvement"
data["performance_level"] = data["Total score"].apply(categorize_performance)
print(data[["gender","Total score","performance_level"]].head())       
# 2. Count number of students in each category
category_counts = data["performance_level"].value_counts()
print("\nNumber of students in each performance category:\n", category_counts)
```

```python
# �Task 9: Top & Bottom Performers Identification .The board wants to recognize excellence and support weak students. 
# 1. Identify top 10 students based on total score. 
top_10_students = data.sort_values(by="Total score", ascending=False).head(10)
print("           === Top 10 Students ===\n")
print(top_10_students[["gender", "math score", "reading score", "writing score", "Total score"]])
# 2. Identify bottom 10 students based on total score. 
bottom_10_students = data.sort_values(by="Total score", ascending=True).head(10)
# 3. Display relevant student details.
print("\n          === Bottom 10 Students ===\n")
print(bottom_10_students[[ "gender", "math score", "reading score", "writing score", "Total score"]])
```

```python
# �Task 10: Insights & Conclusion (Analytical Thinking) .Based on your analysis, answer: 
# 1. Which factor has the strongest impact on student performance?
parental_avg = data.groupby("parental level of education")["Total score"].mean()
strongest_factor = parental_avg.idxmax()
print("1. Factor with strongest impact: Parental Education Level -", strongest_factor)
# 2. Does test preparation significantly improve scores?
test_prep_avg = data.groupby("test preparation course")["Total score"].mean()
print("\n2. Test Preparation Course Impact:")
print(test_prep_avg)
if test_prep_avg["completed"] > test_prep_avg["none"]:
    print("   → Yes, students who completed the test preparation course scored higher on average.")
else:
    print("   → No significant improvement observed from the test preparation course.")
# 3. Which subject shows the highest overall performance?
subject_avg = data[["math score", "reading score", "writing score"]].mean()
highest_subject = subject_avg.idxmax()
print("\n3. Subject with highest overall performance:", highest_subject)
# 4. Provide 2–3 data-driven recommendations for the education board.
print("\n4. Recommendations:")
print("   • Encourage parental involvement and education awareness programs, as parental education strongly impacts student performance.")
print("   • Promote and provide access to test preparation courses to improve student outcomes.")
print("   • Focus additional support on students struggling in math, as it generally shows lower averages compared to reading and writing (if applicable).")
```
